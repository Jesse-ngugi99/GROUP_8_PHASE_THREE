## 3.0 EXPLORATORY DATA ANALYSIS

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
#Loading of the data set
df=pd.read_csv('bigml_59c28831336c6604c800002a.csv')

# preview the data
df.head()
```

```python
#check the columns name and the data types
print(df.info())
```

```python
# checking for missing values
print(df.isnull().sum())
```

```python
# Checking for duplicates
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

df['churn'] = df['churn'].map({'True': 1, 'False': 0})
print(df['churn'].unique())

```python
# Encoding categorical features
df['international plan'] = df['international plan'].map({'yes': 1, 'no': 0})
df['voice mail plan'] = df['voice mail plan'].map({'yes': 1, 'no': 0})
```

```python
# Compute correlation of numerical features with 'churn'
correlation_with_churn = df.corr(numeric_only=True)['churn'].sort_values(ascending=False)

print("Correlation of features with churn:")
print(correlation_with_churn)
```

### 3.1 Visualizing Data Distributions

```python
# Countplot for target variable
sns.countplot(x='churn', data=df, palette='Set2')
plt.title('Churn Distribution')
plt.show()
```

The above plot shows the distribution of the target variable, Churn, which clearly indicates whether a SyriaTel Customerhas left or remained.
- We also see that the dataset is imbalanced with significantly more customers not leaving compared to those who left.
- This imbalance could affect the performance ofmachine learning models particulary the logistic regression

```python
# Histograms for numerical variables
df.select_dtypes(include='number').hist(figsize=(15,10), bins=30, edgecolor='black')
plt.suptitle('Distribution of Numerical Features')
plt.tight_layout()
plt.show()
```

The above plots shows the distribution of all numerical features in our dataset.

- We observe that Customer service calls and number vmail messages have visible outliers
- Total intl charge and total intl calls are skewed distribution and the rest features appear approximately normally distribution.

```python
#box plot
if df['churn'].dtype != 'int64' and df['churn'].dtype != 'int32':
    df['churn'] = df['churn'].astype(int)

# Select numerical features excluding churn
numerical_features = [col for col in df.select_dtypes(include='number').columns if col != 'churn']

# Melt dataframe
df_melted = df.melt(id_vars='churn', value_vars=numerical_features, var_name='Feature', value_name='Value')

# Plot
plt.figure(figsize=(15, 8))
sns.boxplot(data=df_melted, x='Feature', y='Value', hue='churn', palette='Set2')
plt.xticks(rotation=45)
plt.title('Boxplots of Numerical Features by Churn')
plt.tight_layout()
plt.show()
```

The Boxplots helps highlight which numerical features are most associated with churn, aiding in feature selection for building predictive models.
